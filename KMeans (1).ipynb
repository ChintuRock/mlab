```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
initial_centers = np.array([[6.2, 3.2], [6.6, 3.7], [6.5, 3.0]])
```

```python
X = np.array([
    [5.9, 3.2],
    [4.6, 2.9],
    [6.2, 2.8],
    [4.7, 3.2],
    [5.5, 4.2],
    [5.0, 3.0],
    [4.9, 3.1],
    [6.7, 3.1],
    [5.1, 3.8],
    [6.0, 3.0]
])
```

```python
kmeans = KMeans(n_clusters=3, init=initial_centers, n_init=1, random_state=0)
kmeans.fit(X)
center_red_after_one_iteration = kmeans.cluster_centers_[0]
```

```python
kmeans = KMeans(n_clusters=3, init=initial_centers, n_init=2, random_state=0)
kmeans.fit(X)
center_green_after_two_iterations = kmeans.cluster_centers_[1]
```

```python
kmeans = KMeans(n_clusters=3, init=initial_centers, random_state=0)
kmeans.fit(X)
center_blue_converged = kmeans.cluster_centers_[2]
```

```python
n_iterations = kmeans.n_iter_
```

```python
center_red_after_one_iteration = np.round(center_red_after_one_iteration, 3)
center_green_after_two_iterations = np.round(center_green_after_two_iterations, 3)
center_blue_converged = np.round(center_blue_converged, 3)
```

```python
print("1. Center of the 1st cluster (red) after one iteration:", center_red_after_one_iteration)
print("2. Center of the 2nd cluster (green) after two iterations:", center_green_after_two_iterations)
print("3. Center of the 3rd cluster (blue) when the clustering converges:", center_blue_converged)
print("4. Number of iterations required for convergence:",n_iterations)
```

```python
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[kmeans.labels_ == 2, 0], X[kmeans.labels_ == 2, 1], s=100, c='green', label='Cluster 3')
```

```python
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='yellow', label='Centroids')
plt.title('KMeans Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
